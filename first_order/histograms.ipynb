```python
def plot_order(df):
    fig, ax = plt.subplots(2, 2)
    for i, n in enumerate([600, 610, 620, 630]):
        f = df.loc[df.Duty == n]
        frame = f.index.values[0]
        fi = f.loc[frame]
        ax[i//2, i%2].scatter(fi.x, fi.y, c=fi.order_long, s=d.r.mean()/2)
        ax[i//2, i%2].set_title(n)
```

```python
plot_order(dimpled_data.df)
```
