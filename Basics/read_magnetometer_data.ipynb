```python
import pandas as pd
```

```python
data = pd.read_csv("/media/data/Data/GeneralSystem/PhoneMagnetometerReading/Raw Data.csv")
```

```python
data.head()
```

```python
tstr = data.columns[0]
xstr = data.columns[1]
xstr
```

```python
data.plot(x=tstr, y=xstr)
```

```python
t = data[tstr]
Bx = data[xstr]
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.plot(t, Bx)
```

```python
spectrum = np.abs(np.fft.rfft(Bx)**2)
```

```python
freq = np.fft.rfftfreq(len(t), d=t[1]-t[0])
```

```python
%matplotlib auto
plt.plot(freq[1:], spectrum[1:])
plt.xlabel('Frequency (Hz)')
```

```python
plt.subplot(2, 1, 1)
plt.plot(t, Bx)
plt.subplot(2, 1, 2)
plt.plot(freq[1:], spectrum[1:])
```
