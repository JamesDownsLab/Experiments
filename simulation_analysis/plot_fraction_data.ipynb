```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib
```

```python
matplotlib.rcParams.update({'font.size': 22})
```

```python
direc = "/media/NAS/share/James/SimulationDataSep22/plot_data"
```

```python
def plot(densities: [], duties: [], index=0, title=''):
    fig, ax = plt.subplots()
    for density in densities:
        for duty in duties:
            label = duty if index == 0 else density
            try:
                freq = np.loadtxt(f"{direc}/{density}_{duty}_frequencies.txt")
                fraction = np.loadtxt(f"{direc}/{density}_{duty}_mean_fraction.txt")
                fraction_std = np.loadtxt(f"{direc}/{density}_{duty}_mean_fraction_std.txt")
                ax.errorbar(freq, fraction, fraction_std, label=label)
            except:
                pass

    ax.set_xscale('log')
    ax.set_xlabel('Sampling Frequency [Hz]')
    ax.set_ylabel('$v_s/v$')
    ax.legend()
    ax.set_title(title)
    if index == 0:
        plt.savefig(f"/media/NAS/share/James/SimulationDataSep22/Figures/Fractions/{density}.jpg", dpi=300)
    else:
        plt.savefig(f"/media/NAS/share/James/SimulationDataSep22/Figures/Fractions/{duty}.jpg", dpi=300)
```

```python
for density in [80, 82, 85, 87, 90, 92, 95]:
    plot([density], ['3e-4', '3,5e-4', '4e-4'], title=density)

for acc in ['3e-4', '3,5e-4', '4e-4']:
    plot([80, 82, 85, 87, 90, 92, 95], [acc], index=1, title=acc)
```
