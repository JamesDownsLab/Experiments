```python
import cv2
def get_aspect_ratios(df):
    cnt = df[['x', 'y']].values.reshape((-1, 1, 2)).astype('int32')
    x, y, w, h = cv2.boundingRect(cnt)
    aspect = float(w)/h
    return aspect

def get_solidities(df):
    cnt = df[['x', 'y']].values.reshape((-1, 1, 2)).astype('int32')
    area = cv2.contourArea(cnt)
    hull = cv2.convexHull(cnt)
    hull_area = cv2.contourArea(hull)
    solidity = float(area)/hull_area
    return solidity

def get_ratio(df):
    cnt = df[['x', 'y']].values.reshape((-1, 1, 2)).astype('int32')
    (x, y), r = cv2.minEnclosingCircle(cnt)
    circle_area = np.pi * r**2
    hull = cv2.convexHull(cnt)
    hull_area = cv2.contourArea(hull)
    return hull_area/circle_area
```

```python
def hull_vertices_std(df):
    points = df[['x', 'y']].values
    hull = spatial.ConvexHull(points)
    center = points.mean(axis=0)
    vertices = hull.vertices
    vertices = np.append(vertices, vertices[0])
    vertices = points[vertices, :]
    vectors = vertices - center
    angles = np.arctan2(vectors[:, 1], vectors[:, 0])
    angle_diffs = np.abs(angles[1:] - angles[:-1])
    angle_diffs[angle_diffs>np.pi] -= np.pi
    return np.std(angle_diffs)

frames_angles = frames[0].groupby('cluster').apply(hull_vertices_std)
```
